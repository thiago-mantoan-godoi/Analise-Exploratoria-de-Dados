### Descrição do Dataset

O dataset contém registros detalhados da produção de chicotes elétricos em uma fábrica. Cada linha representa um ciclo de produção individual e contém diversas métricas relacionadas ao desempenho da produção, incluindo tempos, quantidades e defeitos. Abaixo está uma descrição das principais colunas do dataset:

- **ID_Produto**: Identificação única de cada produto.
- **Hora_inicio**: Data e hora de início do ciclo de produção.
- **Hora_fim**: Data e hora de término do ciclo de produção.
- **Tempo_ciclo**: Tempo total gasto no ciclo de produção (em minutos).
- **Tempo_espera**: Tempo de espera durante o ciclo de produção (em minutos).
- **Tempo_parada**: Tempo de parada durante o ciclo de produção (em minutos).
- **Tipo_produto**: Categoria do produto produzido.
- **Tempo_total**: Tempo total do ciclo incluindo produção, espera e paradas (em minutos).
- **Dia_semana**: Dia da semana em que o ciclo de produção ocorreu.
- **Quantidade_defeitos**: Número de defeitos identificados no produto.
- **Descricao_defeitos**: Descrição dos defeitos encontrados no produto.
- **Quantidade_cabos**: Número de cabos usados no chicote elétrico.
- **Quantidade_terminais**: Número de terminais usados no chicote elétrico.
- **Quantidade_conectores**: Número de conectores usados no chicote elétrico.
- **Quantidade_clips**: Número de clips usados no chicote elétrico.

O dataset abrange um total de 148.585 registros, proporcionando uma visão abrangente sobre a eficiência da produção, a ocorrência de defeitos e os tempos associados a cada ciclo de produção.

### Resultados da Análise Exploratória

#### 1. Distribuição do Tempo de Ciclo

A maioria dos tempos de ciclo está concentrada em torno de 2 minutos, com uma distribuição que se assemelha a uma distribuição normal. Existem poucos valores atípicos, mas a maioria dos ciclos de produção ocorre em menos de 4 minutos.

#### 2. Distribuição do Tempo de Parada

A distribuição do tempo de parada mostra uma concentração significativa em torno de 0 minutos, indicando que muitas operações não enfrentaram paradas. No entanto, há um número considerável de paradas de até 35 minutos.

#### 3. Análise de Defeitos

Os defeitos mais comuns são:

1. Mecanismo de travamento quebrado.
2. Componente não fixado corretamente.
3. Conector de encaixe danificado.
4. Isolamento derretido devido a superaquecimento.
5. Problema de contato elétrico intermitente.

### Insights

1. **Eficiência da Produção**: O tempo de ciclo está bem controlado, com a maioria das operações sendo concluídas em 2 minutos. Isso sugere um processo de produção eficiente.
   
2. **Paradas na Produção**: A alta frequência de paradas de até 35 minutos indica a necessidade de investigar e mitigar as causas dessas interrupções para melhorar a eficiência operacional.

3. **Defeitos Comuns**: Identificar e mitigar os defeitos mais comuns pode melhorar significativamente a qualidade do produto e reduzir retrabalhos. A manutenção preventiva e treinamentos específicos podem ser implementados para reduzir esses defeitos.
